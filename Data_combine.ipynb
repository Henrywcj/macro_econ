```python
import numpy as np
import pandas as pd
from datetime import timedelta

def date_adj(x):
    return x - timedelta(days=1)
```

```python
US_M1_df = pd.read_csv("./M1.csv")
US_M2_df = pd.read_csv("./M2.csv")
TB3MS_df = pd.read_csv("./TB3MS.csv")
TB6MS_df = pd.read_csv("./TB6MS.csv")
US_yield_curve = pd.read_excel("./yield curve.xlsx")
US_gdp_df = pd.read_csv("./US_gdp_Q.csv")
US_recession_df = pd.read_csv("./US_recession_Q.csv")

DJIA_df = pd.read_csv("./DJIA_monthly.csv")
NYSE_df = pd.read_csv("./NYSE_monthly.csv")
SP500_df = pd.read_csv("./S&P500_monthly.csv")
```

```python
US_M1_df['DATE'] = pd.to_datetime(US_M1_df['DATE'])
```

```python
US_M1_df.set_index('DATE')['M1SL']
```

```python
US_M1_df.set_index('DATE')['M1SL'].shift(1)
```

```python
# adjusting for lag (1)
US_M1_df['DATE'] = pd.to_datetime(US_M1_df['DATE'])
US_M1_Q = US_M1_df.set_index('DATE')['M1SL'].resample('1Q').last()
US_M2_df['DATE'] = pd.to_datetime(US_M2_df['DATE'])
US_M2_Q = US_M2_df.set_index('DATE')['M2SL'].resample('1Q').last()
```

```python
# Interest Rate and spreads are quarterly average levels 
TB3MS_df['DATE'] = pd.to_datetime(TB3MS_df['DATE'])
TB3MS_Q = TB3MS_df.set_index('DATE')['TB3MS'].resample('1Q').mean()
TB6MS_df['DATE'] = pd.to_datetime(TB6MS_df['DATE'])
TB6MS_Q = TB6MS_df.set_index('DATE')['TB6MS'].resample('1Q').mean()
US_yield_curve['Date'] = pd.to_datetime(US_yield_curve['Date'])
US_yield_curve = US_yield_curve.set_index('Date').resample('1Q').mean()
```

```python
US_gdp_df['DATE'] = pd.to_datetime(US_gdp_df['DATE'])
US_gdp_df['DATE'] = US_gdp_df['DATE'].apply(lambda x: date_adj(x))
US_gdp_Q = US_gdp_df.set_index('DATE')['GDPC1'].resample('1Q').last()
US_recession_df['DATE'] = pd.to_datetime(US_recession_df['DATE'])
US_recession_df['DATE'] = US_recession_df['DATE'].apply(lambda x: date_adj(x))
US_recession_Q = US_recession_df.set_index('DATE')['USRECQ'].resample('1Q').last()
```

```python
DJIA_df['Date'] = pd.to_datetime(DJIA_df['Date'])
DJIA_df['Date'] = DJIA_df['Date'].apply(lambda x: date_adj(x))
DJIA_Q = DJIA_df.set_index('Date').resample('1Q').last()
DJIA_Q.columns = DJIA_Q.columns.to_series().apply(lambda x: 'DJIA_'+x)

NYSE_df['Date'] = pd.to_datetime(NYSE_df['Date'])
NYSE_df['Date'] = NYSE_df['Date'].apply(lambda x: date_adj(x))
NYSE_Q = DJIA_df.set_index('Date').resample('1Q').last()
NYSE_Q.columns = NYSE_Q.columns.to_series().apply(lambda x: 'NYSE_'+x)

SP500_df['Date'] = pd.to_datetime(SP500_df['Date'])
SP500_df['Date'] = SP500_df['Date'].apply(lambda x: date_adj(x))
SP500_Q = DJIA_df.set_index('Date').resample('1Q').last()
SP500_Q.columns = SP500_Q.columns.to_series().apply(lambda x: 'SP500_'+x)
```

```python
comb_df = pd.concat([US_M1_Q, US_M2_Q, TB3MS_Q, TB6MS_Q, US_gdp_Q, US_recession_Q, US_yield_curve, DJIA_Q, NYSE_Q, SP500_Q], axis=1)
```

```python
comb_df.to_csv('./comb_df.csv')
```
